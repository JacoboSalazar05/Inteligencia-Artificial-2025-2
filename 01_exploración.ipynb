```python
# ===========================
# 01 - Exploración.ipynb
# ===========================

# --- Celda 1: Imports ---
import pandas as pd
import matplotlib.pyplot as plt

# Configuración de visualización
pd.set_option('display.max_columns', 50)


# --- Celda 2: Cargar el dataset ---
# Ajusta la ruta según donde tengas el archivo
path = "train.csv"
df = pd.read_csv(path)

print("Shape:", df.shape)
df.head()


# --- Celda 3: Información general ---
print("\n--- Info del DataFrame ---")
print(df.info())

print("\n--- Primeras filas ---")
display(df.head())

print("\n--- Descripción de variables numéricas ---")
display(df.describe())


# --- Celda 4: Número de estudiantes por programa académico ---
print("\n--- Top 15 programas académicos con más estudiantes ---")
programas = df['E_PRGM_ACADEMICO'].value_counts().head(15)
display(programas)

programas.plot(kind='bar', figsize=(10,4), title="Estudiantes por programa académico (Top 15)")
plt.show()


# --- Celda 5: Número de estudiantes por departamento ---
print("\n--- Top 15 departamentos con más estudiantes ---")
departamentos = df['E_PRGM_DEPARTAMENTO'].value_counts().head(15)
display(departamentos)

departamentos.plot(kind='bar', figsize=(10,4), title="Estudiantes por departamento (Top 15)")
plt.show()


# --- Celda 6: Distribución del rendimiento global ---
print("\n--- Distribución de RENDIMIENTO_GLOBAL ---")
rendimiento = df['RENDIMIENTO_GLOBAL'].value_counts(normalize=True) * 100
display(rendimiento)

rendimiento.plot(kind='bar', figsize=(6,4), title="Distribución del rendimiento global (%)")
plt.ylabel("Porcentaje")
plt.show()


# --- Celda 7: Cruce entre departamento y rendimiento global ---
print("\n--- Rendimiento por departamento (primeros 10) ---")
cruce = pd.crosstab(df['E_PRGM_DEPARTAMENTO'], df['RENDIMIENTO_GLOBAL'])
display(cruce.head(10))


# --- Celda 8: Valores nulos ---
print("\n--- Porcentaje de valores nulos por columna ---")
nulos = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
display(nulos.head(10))
```

```python
# ============================================
# 01 - Exploración de datos
# ============================================

# Paso 1. Importar librerías
import pandas as pd

# Paso 2. Cargar el dataset
# Asegúrate de tener "train.csv" en la misma carpeta del notebook
df = pd.read_csv("train.csv")

# Mostrar las primeras filas
print("Primeras filas del dataset:")
display(df.head())


# Paso 3. Información general del dataset
print("\nInformación general del dataset:")
df.info()


# Paso 4. Cantidad de estudiantes por rendimiento global
print("\nCantidad de estudiantes por rendimiento global:")
rendimiento = (
    df["RENDIMIENTO_GLOBAL"]
    .value_counts()
    .sort_index()  # <-- ordena alfabéticamente para consistencia
)
print(rendimiento)


# Paso 5. Cantidad de estudiantes por departamento (top 10)
print("\nCantidad de estudiantes por departamento (top 10):")
departamentos = (
    df["E_PRGM_DEPARTAMENTO"]
    .value_counts()
    .sort_values(ascending=False)  # orden descendente por conteo
    .sort_index(ascending=True)    # en caso de empate, orden alfabético
    .head(10)
)
print(departamentos)


# Paso 6. Cantidad de estudiantes por programa académico (top 10)
print("\nCantidad de estudiantes por programa académico (top 10):")
programas = (
    df["E_PRGM_ACADEMICO"]
    .value_counts()
    .sort_values(ascending=False)  # orden descendente por conteo
    .sort_index(ascending=True)    # en caso de empate, orden alfabético
    .head(10)
)
print(programas)
```
